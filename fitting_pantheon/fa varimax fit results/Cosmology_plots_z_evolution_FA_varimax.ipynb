```python
flat_samples_exp=[]
for j in range(18):
    flat_samples_exp.append(np.loadtxt("./chains_evol_w_z_FA_varimax/chain_int"+str(j)+"_FA_varimax.txt"))
```

```python
z_values=np.array([0.01725671293371204,0.029638071875199994,0.045190399517381394,0.08668939944370001,
0.12697050000000001,0.1532592307692308,0.18017504545454543,0.2036151860465117,0.23034628888888878,
0.25590036170212765,0.2846988095238095,0.3169796976744187,0.3552084186046511,0.4133853658536586,
0.4964214285714287,0.5954957446808513,0.7441477272727272,0.9451444444444442])
```

```python
alpha_values=[]
alpha_error_up=[]
alpha_error_down=[]
beta_values=[]
beta_error_up=[]
beta_error_down=[]
sig_values=[]
sig_error_up=[]
sig_error_down=[]

for j in range(18):
    percentile_alpha=np.percentile(flat_samples_exp[j][:,0],[16,50,84])
    alpha_values.append(percentile_alpha[1]) 
    alpha_error_up.append(np.diff(percentile_alpha)[1])
    alpha_error_down.append(np.diff(percentile_alpha)[0])
    
    percentile_beta=np.percentile(flat_samples_exp[j][:,1],[16,50,84])
    beta_values.append(percentile_beta[1]) 
    beta_error_up.append(np.diff(percentile_beta)[1])
    beta_error_down.append(np.diff(percentile_beta)[0])

    percentile_sig=np.percentile(flat_samples_exp[j][:,2],[16,50,84])
    sig_values.append(percentile_sig[1]) 
    sig_error_up.append(np.diff(percentile_sig)[1])
    sig_error_down.append(np.diff(percentile_sig)[0])
```

```python
fig, axs= plt.subplots(3,1,figsize=(6,12),sharex=True)

axs[0].errorbar(z_values[1:],alpha_values[1:],yerr=[alpha_error_up[1:],alpha_error_down[1:]],ls="none",marker="o",markersize=6)
axs[0].axhline(y=0.85797,c="k")
axs[0].set_ylabel(r"$\alpha$")
axs[1].errorbar(z_values[1:],beta_values[1:],yerr=[beta_error_up[1:],beta_error_down[1:]],ls="none",marker="o",markersize=6)
axs[1].axhline(y=2.40310,c="k")
axs[1].set_ylabel(r"$\beta$")
axs[2].errorbar(z_values[1:],sig_values[1:],yerr=[sig_error_up[1:],sig_error_down[1:]],ls="none",marker="o",markersize=6,c="C0")
axs[2].axhline(y=0.14366,c="k")
axs[2].set_xlabel("z")
axs[2].set_ylabel(r"$\sigma_{int}$")

plt.tight_layout()
plt.savefig('./parameters_evol_w_z_FA_varimax.pdf');
```
