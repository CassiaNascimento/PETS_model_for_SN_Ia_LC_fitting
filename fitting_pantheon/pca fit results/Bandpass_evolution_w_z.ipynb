```python
import sncosmo
import matplotlib.pyplot as plt
import numpy as np
```

```python
x=np.linspace(2000,9000,200)
```

```python
dx=(9000-2000)/200
```

```python
def z_shift_bandpass(z):
    y=x*sncosmo.get_bandpass("standard::b")(x*(1+z))
    return y
```

```python
def z_shift_bandpass_times_z(z):
    y=(1+z)*x*sncosmo.get_bandpass("standard::b")(x*(1+z))
    return y
```

```python
plt.plot(x,z_shift_bandpass(0),label="z=0")
plt.plot(x,z_shift_bandpass(0.01),label="z=0.01")
plt.plot(x,z_shift_bandpass(0.1),label="z=0.1")
plt.plot(x,z_shift_bandpass(0.2),label="z=0.2")
plt.plot(x,z_shift_bandpass(0.3),label="z=0.3")
plt.plot(x,z_shift_bandpass(0.4),label="z=0.4")
plt.plot(x,z_shift_bandpass(0.5),label="z=0.5")
plt.legend();
```

```python
plt.plot(x,z_shift_bandpass_times_z(0),label="z=0")
plt.plot(x,z_shift_bandpass_times_z(0.01),label="z=0.01")
plt.plot(x,z_shift_bandpass_times_z(0.1),label="z=0.1")
plt.plot(x,z_shift_bandpass_times_z(0.2),label="z=0.2")
plt.plot(x,z_shift_bandpass_times_z(0.3),label="z=0.3")
plt.plot(x,z_shift_bandpass_times_z(0.4),label="z=0.4")
plt.plot(x,z_shift_bandpass_times_z(0.5),label="z=0.5")
plt.legend();
```

```python
salt2=sncosmo.Model(source="salt2")
```

```python
plt.plot(x,salt2.flux(0,x));
```

```python
def integral(z):
    return  np.sum(salt2.flux(0,x)*(1+z)*x*sncosmo.get_bandpass("standard::b")(x*(1+z))*dx)
```

```python
def integral2(z):
    return  np.sum(salt2.flux(0,x)*x*sncosmo.get_bandpass("standard::b")(x)*dx)
```

```python
plt.plot(np.linspace(0.,1.,20),[integral(z) for z in np.linspace(0.,1.,20)]);
plt.plot(np.linspace(0.,1.,20),[integral2(z) for z in np.linspace(0.,1.,20)])
```
